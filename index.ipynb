## Define the interval bounds which contain 99% of the observations   

> **Hint**: Recall the empirical rule related to $3\sigma$.

```python
# Perform any calculations necessary here
sigma=yield_pd['0'].std()
sigma
mu=np.mean(yield_pd['0'])
mu
```

```python
# Write your answer here 
ans=mu-3*sigma
ans
```

## Compute and interpret the z-score for a tree yielding 35 pounds of apples

```python
# Calculate z
yield_35=35
z=(yield_35-mu)/sigma
z
```

```python
# Interpret the result
"""the tree yielding 35 pounds of apples is 1 standard deviation below the mean"""
"""it is therefore lower than the average"""
```

## Suppose a tree has a z-score of 1.85. Interpret this z-score. What is the yield of this tree?

```python
# Interpret the z score
"""the tree is 1.85 standard deviation above the mean"""
```

```python
# Calculate yield
z_score=1.85
#1.85=(x-mu)/sigma
yield_tree=mu-(z_score*sigma)
yield_tree
```

```python
# What is the yield ?
print(round(yield_tree,0))
```

##  Convert each tree’s yield to a z-score so the new variable is the “z-score for weight”

The units are still the apple trees. For the data set of all z-scores:

* What is the shape? 
* The mean? 
* The standard deviation?

```python
# Give your solution here 
correct_z=yield_pd['0'] +4

tree_z=(correct_z-mu)/sigma
x_mean=np.mean(tree_z)
x_std=np.std(tree_z)
print(x_mean,x_std)
sns.histplot(tree_z,kde=True)

```
